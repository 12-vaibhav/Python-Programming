<h2>What is an object in python?</h2>

<ul>
    <li>An object is a runtime entity that has: </li>
    <li>Identity: unique memory reference (id())</li>
    <li>Type: what kind of object it is (type())</li>
    <li>Value / State: actual data stored</li>
</ul>

<br>

<p>Example: </p>

```python
a = 12

#print the id of variable a
print(id(a))

#print the type of variable a
print(type(a))

#print the value of variable a
print(a)
```

<p>Internally: </p>
<ul>
    <li>12 is an object of a class int</li>
    <li>int itself is also an object (a class object)</li>
</ul>

<h2>How objects are Called / Accessed in Python?</h2>
<h3>Dot operator (.) is attribute access</h3>

<p>When we write: </p>

<br>

```python
"hello world".upper()
```

<p>Python Interprets: </p>

<ul>
    <li>"hello world" : object</li>
    <li>upper : attribute (method) stored inside the class string</li>
</ul>

<p>So internally it becomes :</p>

<p>str.upper("hello world") <br><br> Meaning: </p>

<p>Methods are just functions stored inside a class and Python passes the object automatically.</p>

<h2>Method (belongs to an object's class)</h2>
<p>Example: </p>

```python
"codewithvaibhav".upper()
```

<h2>Internal Meaning: </h2>
<p>Python does: </p>

<ol>
    <li>Look at object type: string</li>
    <li>Find upper inside string</li>
    <li>Bind it with the object "codewithvaibhav"</li>
    <li>Then call it</li>
</ol>

<p>Equivalent internal call: <br>
str.upper("codewithvaibhav") <br>
So, method = function that becomes bound to an object.
</p>
