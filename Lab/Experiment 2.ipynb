<h2>Write a program where the radius is taken as input to compute the area of a circle.</h2>

```python
import math
RadiusInput = input("Enter the radius of the circle: ")
Radius = float(RadiusInput)

AreaOfCircle = math.pi * (Radius **2)

print("Radius: ", RadiusInput)
print(f"The area of circle: {AreaOfCircle:.2f}")
```

<h2>Write a program to compute the length of the hypotenuse (c) of a right triangle using Pythagoras theorem.</h2>

```python
import math

a = float(input("Enter the length of side a"))
b = float(input("Enter the length of side b"))

c = math.sqrt(a**2 + b**2)

print("Length of side a", a)
print("Length of side b", b)

print("The length of the hypotenuse (c) is", c)
```
