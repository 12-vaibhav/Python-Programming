<h2><b>Question: </b>Write a program to compute the length of the hypotenuse (c) of a right triangle using Pythagoras theorem.</h2>
<h2>Code: </h2>

```python
import math

a = float(input("Enter the length of side a"))
b = float(input("Enter the length of side b"))

c = math.sqrt(a**2 + b**2)

print("Length of side a", a)
print("Length of side b", b)

print("The length of the hypotenuse (c) is", c)
```

<h2><b>Question: </b>Write a program to convert given seconds into hours, minutes and remaining seconds.</h2>
<h2>Code: </h2>

```python
# Taking input for seconds to convert it into Hours, Minutes and remaining seconds
seconds = int(input("Enter the seconds"))

Hours = seconds / 3600

Minutes = seconds / 60

remainingSeconds = seconds % 60

print(f"{Hours:.1f} Hours,", f"{Minutes:.0f} Minutes,", remainingSeconds)
```

<h2><b>Question: </b>Write a program to print truth table for bitwise operators (&, | and ^ operators).</h2>
<h2>Code: </h2>

```python
# Truth table for program for Bitwise Operator
print(f"{'A':<3} {'B':<3} | {'A & B':<7} {'A | B':<7} {'A ^ B':<7}")
print("-" * 35)

# Itherating through all binary combinations
for a in [0, 1]:
    for b in [0, 1]:
        bitwiseAnd = a & b
        bitwiseOr = a | b
        bitwiseXor = a ^ b

        print(f"{a:<3} {b:<3} | {bitwiseAnd:<7} {bitwiseOr:<7} {bitwiseXor:<7}")
```

<h2><b>Question: </b>Write a program to find left shift and right shift values of a given number.</h2>
<h2>Code: </h2>

```python
# Taking input for number and shift
number = int(input("Enter a number "))
shift = int(input("Enter number of position to shift "))

# Performing Bitwise Shift
leftShift = number << shift
rightShift = number >> shift

print(f"Left Shift ({number} << {shift}) : {leftShift}")
print(f"Right Shift ({number} >> {shift}) : {rightShift}")
```
