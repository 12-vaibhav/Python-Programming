<h2>Declare these variables (x, y and z) as integers. Assign a value of 9 to x. Assign a value of 7 to y, perform addition, multiplication, division and subtraction on these two variables and print out the result.</h2>

```python
x, y = 9, 7
print(f"Adding two integers {x} and {y} we get:", x + y)
print(f"Multiplying two integers {x} and {y} we get:", x * y)
print(f"Dividing two integers {x} and {y} we get:", x / y)
print(f"Subtracting two integers {x} and {y} we get:", x - y)
```

<h2>Write a program where the radius is taken as input to compute the area of a circle.</h2>

```python
import math
RadiusInput = input("Enter the radius of the circle: ")
Radius = float(RadiusInput)

AreaOfCircle = math.pi * (Radius **2)

print("Radius: ", RadiusInput)
print("The area of circle: ", AreaOfCircle)
```
