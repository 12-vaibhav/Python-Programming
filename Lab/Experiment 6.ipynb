<h2><b>Question: </b>Scan n values in range 0-3 and print the number of times each value has occurred.</h2>
<h2>Code: </h2>

```python
counts = [0] * 4

n = int(input("Enter the number of values (n) "))

print(f"Enter {n} values between 0 and 3 ")
      
for _ in range (n):
    value = int(input())

    if 0 <= value <= 3:
        counts[value] += 1
    else:
        print(f"Value {value} is out of range!!!")

for i in range(4):
    print(f"Value {i} occured {counts[i]} times.")
```

<h2><b>Question: </b>Create a tuple to store n numeric values and find the average of all values.</h2>
<h2>Code: </h2>

```python
# Inputting the number of values
n = int(input("Enter the number of values "))

# Collecting values in a list
numberList = []
for i in range(n):
    value = float(input(f"Enter value {i+1} "))
    numberList.append(value)

# Coverting the list into a tuple
numberTuple = tuple(numberList)

# Calculating the average
if len(numberTuple) > 0:
    average = sum(numberTuple) / len(numberTuple)
    print(f"\nTuple: {numberTuple}")
    print(f"Average: {average}")
else:
    print("No values are entered.")
```

<h2><b>Question: </b>Write a program to input a list of scores for N students in a list data type. Find the score of the runner-up and print the output.</h2>
<h2>Code: </h2>

```python
n = int(input("Enter number of students "))
scores = list(map(int, input("Enter all scores separated by spaces ").split()))

uniqueScores = sorted(list(set(scores)))

if len(uniqueScores) < 2:
    print("Not enough unique scores to find a runner-up.")
else:
    print("Runner-Up score", uniqueScores[-2])
```

<h2><b>Question: </b>Store details of n movies in a dictionary by taking input from the user. Each movie must store details like name, year, director name, production cost, collection made (earning) & perform the following: </h2>

<ol type = "a">
    <li>Print all movie detaills.</li>
    <li>Display name of movies released before 2015.</li>
    <li>Print movies that made a profit.</li>
    <li>Print movies directed by a particular director.</li>
</ol>

<h2>Code:</h2>

```python
# Initializing the dictionary and getting the number of movies
n = int(input("Enter the number of movies "))
movies = {}

# Inputting data from the user
for i in range(n):
    print(f"\nEnter details for the Movie {i + 1} ")
    name = input("Name ")
    year = int(input("Year of Release "))
    director = input("Director Name ")
    cost = float(input("Production cost "))
    earning = float(input("Collection Made (earning) "))

    # Storing in the dictionary
    movies[name] = {
        "year": year,
        "director": director,
        "cost": cost,
        "earning": earning
    }

# Printing all movies details
print("\n --- All Movies Details ---")
for name, info in movies.items():
    print(f"Name: {name} | Year: {info['year']} | Director: {info['director']} | Cost: {info['earning']}")

# Displaying the name of movies released before 2015
print("\n --- Movies Released Before 2015 ---")
foundOld = False
for name, info in movies.items():
    if info['year'] < 2015:
        print(name)
        foundOld = True
if not foundOld: print("None found.")

# Printing movies that made a profit
print("\n--- Movies that Made a Profit ---")
foundProfit = False
for name, info in movies.items():
    if info['earning'] > info['cost']:
        profit = info['earning'] - info['cost']
        print(f"{name} (Profit: {profit})")
        foundProfit = True
if not foundProfit: print("No profitable movies found.")

# Printing movies directed by a particular director
searchDirector = input("\nEnter Director Name to search ")
print(f"--- Movies Directed by {searchDirector} ---")
foundDirector = False
for name, info in movies.items():
    if info['director'].lower() == searchDirector.lower():
        print(name)
        foundDirector = True
if not foundDirector: print("No movies found for this director.")
```
