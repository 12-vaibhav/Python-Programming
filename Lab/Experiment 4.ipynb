<h2><b>Question: </b>Find a factorial of given number.</h1>
<h2>Code:</h2>

```python
# Taking input for a number
number = int(input("Enter a number "))
factorial = 1

# Checking for factorial number, also using for loop
if number < 0:
    print("Factorial is not defined for negative numbers")
elif number == 0:
    print(f"The factorial of {number} is 1")
else:
    for i in range(1, number + 1):
        factorial *= i

    print(f"The factorial of {number} is {factorial}.")
```

<h2><b>Question: </b>Find whether the given number is Armstrong number.</h2>
<h2>Code:</h2>

```python
number = int(input("Enter a number "))

temporaryNumber = number
sumValue = 0

order = len(str(number))

while temporaryNumber > 0:
    digit = temporaryNumber % 10
    sumValue += digit ** order
    temporaryNumber //= 10

if number == sumValue:
    print(f"{number} is an Armstrong number")
else:
    print(f"{number} is not an Armstrong number")
```

<h2><b>Question: </b>Print fibonacci series upto given term.</h2>
<h2>Code: </h2>

```python
# Taking input for series of Fibonacci numbers
series = int(input("Enter the number to print Fibonacci series "))

previous2 = 0
previous1 = 1
print(previous2)
print(previous1)

# Using for loop to execute fibonacci series upto given range from user
for fibonacci in range(series):
    newFibonacci = previous1 + previous2
    print(newFibonacci)
    previous2 = previous1
    previous1 = newFibonacci
```

<h2><b>Question: </b>Write a program to find if given number is prime number or not.</h2>
<h2>Code: </h2>

```python
number = int(input("Enter a number: "))

isPrime = True

if number <= 1:
    isPrime = False
else:
    for i in range (2, number):
        if number % i == 0:
            isPrime = False
            break

if isPrime:
    print(f"{number} is a prime number")
else:
    print(f"{number} is not a prime number")
```

<h2><b>Question: </b>Check whether given number is pallindrome or not.</h2>
<h2>Code: </h2>

```python
number = int(input("Enter a number: "))

originalNumber = number
reversedNumber = 0

while number > 0:
    digit = number % 10
    reversedNumber = (reversedNumber * 10) + digit
    number = number // 10

if originalNumber == reversedNumber:
    print(f"{originalNumber} is a pallindrome number.")
else:
    print(f"{originalNumber} is not a pallindrome number.")
```

<h2><b>Question: </b>Write a program to print sum of digits.</h2>
<h2>Code: </h2>

```python
number = int(input("Enter a number: "))
sumDigits = 0

while number > 0:
    digit = number % 10
    sumDigits += digit
    number //= 10

print(f"The sum of digits is {sumDigits}")
```

<h2><b>Question: </b>Count and print all numbers divisible by 5 or 7 between 1 to 100.</h2>
<h2>Code: </h2>

```python
count = 0

print("Numbers divisible by 5 or 7")

for i in range(1, 101):
    if i % 5 == 0 or i % 7 == 0:
        print(i, end = " ")
        count += 1

print(f"\nTotal count: {count}")
```

<h2>Code: </h2>

```python
text = str(input("Enter your name in lowercase: "))
convertedText = ""

for char in text:
    convertedText += text.upper()
    break

print(convertedText)
```

<h2><b>Question: </b>Print all Prime numbers between 1 to 100.</h2>
<h2>Code: </h2>

```python
print("Prime numbers between 2 and 100")
for number in range (2, 101):
    isPrime = True
    for i in range (2, int (number ** 0.5) + 1):
        if number % i == 0:
            isPrime = False
            break

    if isPrime:
        print(number, end = " ")
```
