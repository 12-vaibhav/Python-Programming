<h2><b>Question: </b>Create numpy array to find sum of all elements in an array?</h2>
<h2><b>Code: </b></h2>

```python
import numpy as np

array = np.array([7, 12, 9, 4, 11, 25, 15])
print("Array: ", array)
print("Sum of all elements: ", np.sum(array))
```

<h2><b>Question: </b>Create numpy array of (3, 3) dimension. Now find sum of all rows and columns individually. Also find 2<sup>nd</sup> maximum element in the array.</h2>
<h2><b>Code: </b></h2>

```python
matrix_3x3 = np.array([
    [9, 12, 15],
    [7, 12, 9],
    [4, 11, 25]
])

SumOfRows = np.sum(matrix_3x3, axis=1)
SumOfColumns = np.sum(matrix_3x3, axis=0)
SecondMaximum = np.unique(matrix_3x3.flatten()) [-2]

print("3x3 Matrix ", matrix_3x3)
print("Sum of Rows: ", SumOfRows)
print("Sum of Columns: ", SumOfColumns)
print("Second Maximum element: ", SecondMaximum)
```

<h2><b>Question: </b>Perform Matrix multiplication of any 2 n*n matrices.</h2>
<h2><b>Code: </b></h2>

```python
MatrixA = np.array([
    [7, 12, 9],
    [11, 20, 25],
    [3, 9, 12]
])

MatrixB = np.array([
    [7, 12, 9],
    [11, 20, 25],
    [3, 9, 12]
])

productOfMatrix = MatrixA @ MatrixB

print("Matrix A: ", MatrixA)

print("Matrix B: ", MatrixB)

print("MatrixA x MatrixB", productOfMatrix)
```

<h2><b>Question: </b>Write a Pandas program to get the powers of an array values element-wise.</h2>
<h2>Code: </h2>

```python
import pandas as pd

baseValue = pd.Series([0, 1, 2, 3, 4, 5, 6], name = "base")
powerValue = pd.Series([0, 1, 2, 3, 4, 5, 6], name = "power")
poweredValues = baseValue.pow(powerValue)

print("Base Values: ", baseValue)
print("Power Values: ", poweredValues)

print("Powers of an array values element-wise: ", poweredValues)
```
