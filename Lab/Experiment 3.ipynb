<h2><b>Question: </b>Check whether the quadratic equation has real roots or imaginary roots. Display the roots.</h2>
<h2>Code:</h2>

```python
import cmath

# Taking input for coefficients
a = float(input("Enter a: "))
b = float(input("Enter b: "))
c = float(input("Enter c: "))

if a == 0:
    print("This is not a quadratic equation (a cannot be 0).")
else:
    d = (b**2) - (4*a*c)
    root1 = (-b - cmath.sqrt(d)) / (2*a)
    root2 = (-b + cmath.sqrt(d)) / (2*a)

    if d > 0:
        print("The roots are Real and Distinct.")
    elif d == 0:
        print("The roots are Real and Equal.")
    else:
        print("The roots are Imaginary.")

    print(f"Root 1: {root1}")
    print(f"Root 2: {root2}")
```

<h2><b>Question: </b>Find whether a given year is leap year or not.</h2>
<h2>Code:</h2>

```python
import calendar

year = int(input("Enter the year, to check whether a given year is leap year or not "))

if calendar.isleap(year):
    print(f"The {year} is a leap year.")
else:
    print(f"The {year} is not a leap year.")
```

<h2><b>Question: </b>Write a program which takes any date as input and display next date of the calendar.</h2>
<h2>Code: </h2>

```python
from datetime import date, timedelta

inputDate = input("Enter date (Day, Month, Year): ")


Date, Month, Year = map(int, inputDate.split(","))

currentDate = date(Year, Month, Date)

nextDate = currentDate + timedelta(days=1)

print(f"Day = {nextDate.day} Month = {nextDate.month} Year = {nextDate.year}")
```

<h2><b>Question: </b>Print the grade sheet of a student for the given range of CGPA. Scan marks of five subjects and calculate the percentage.</h2>
<h2>Code: </h2>

```python
def gradeSheet(cgpa):
    if cgpa >= 9.0: return "O (Outstanding)"
    elif cgpa >= 8.0: return "A+ (Excellent)"
    elif cgpa >= 7.0: return "A (Very Good)"
    elif cgpa >= 6.0: return "B+ (Good)"
    elif cgpa >= 5.0: return "B (Above Average)"
    elif cgpa >= 4.0: return "C (Pass)"
    else: return "F (Fail)"

subjects = ["WebTechnologies", "Programming in C", "Linux Lab", "Managing Self", "Env. Sustainability"]

# Taking input for all marks at once
marksInput = input("Enter all 5 marks separated by commas or spaces: ")

# Cleaning the input: replacing commas with spaces, then split into a list
marksStringList = marksInput.replace(',', ' ').split()

# Converting the strings to floats
marks = [float(m) for m in marksStringList]

if len(marks) != 5:
    print(f"Error: You entered {len(marks)} marks, but we need 5.")
else:
    # Calculations
    totalMarks = sum(marks)
    percentage = (totalMarks / 50) * 100 
    cgpa = percentage / 9.5

    # Displaying the Grade Sheet
    print("\n" + "="*40)
    print(f"{'STUDENT GRADE SHEET':^40}")
    print("="*40)
    for sub, m in zip(subjects, marks):
        print(f"{sub:25}: {m}")
    print("-" * 40)
    print(f"Total Marks    : {totalMarks}/50")
    print(f"Percentage     : {percentage:.2f}%")
    print(f"CGPA           : {cgpa:.2f}")
    print(f"Final Grade    : {gradeSheet(cgpa)}")
    print("="*40)
```
