<h2><b>Question: </b>Write a program to count and display the number of capital letters in a given string.</h2>
<h2>Code: </h2>

```python
count = 0
inputString = input("Enter your Name ")

for char in inputString:
    if char.isupper():
        count += 1

print(inputString)
print(f"Number of capital letters {count}")
```

<h2><b>Question: </b>Count total number of vowels in a given string.</h2>
<h2>Code: </h2>

```python
count = 0

inputString = input("Enter your Name ")
Vowels = ["a", "e", "i", "o", "u", "A", "E", "I", "O", "U"]

for char in inputString:
    if char in Vowels:
        count += 1

print(f"Number of Vowels in {inputString} is {count}")
```

<h2><b>Question: </b>Given a string containing both upper and lower-case alphabets. Write a Python program to count the number of occurrences of each alphabet and display the same.</h2>
<h2>Code: </h2>

```python
from collections import Counter
import string

inputString = input("Enter a sentence to count the number of occurrences of each alphabet ")

alphabets = ''.join(char.lower() for char in inputString if char.isalpha())

frequency = Counter(alphabets)

for letter in sorted(frequency):
    print(f"'{letter}': {frequency[letter]}")
```

<h2><b>Question: </b>Program to count the number of unique words in a given sentence using sets.</h2>
<h2>Code: </h2>

```python
inputString = input("Enter a sentence ")

words = inputString.lower().split()

uniqueWords = set(words)

count = len(uniqueWords)

print(f"The number of unique words is {count}.")
```

<h2><b>Question: </b>Create two sets s1 and s2 of n fruits each by taking input from user and find: </h2>
<ol type="a">
    <li>Fruits which are in both sets s1 and s2</li>
    <li>Fruits only in s1 but not in s2</li>
</ol>

<h2>Code: </h2>

```python
n = int(input("Enter number of fruits: "))

Set1 = {input(f"Set1 fruits {i+1} ")for i in range(n)}
Set2 = {input(f"Set2 fruits {i+1} ")for i in range(n)}

commonFruits = Set1 & Set2

print(Set1, Set2)
print("\nCommon fruits ", commonFruits if commonFruits else "None")
```
