```python
class Employee:
    # Class Variable
    CompanyName = "ABC PVT LTD"

    # Instance Variables
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary
    
    # Using Instance method to display employee data
    def show_Details(self):
        print(f"Name: {self.name} \n Salary: {self.salary} \n Company: {self.CompanyName}")

    @classmethod
    def change_company(cls, new_name):
        cls.CompanyName = new_name

# Creating two objects
employee1 = Employee("Alice", 50000)
employee2 = Employee("Bob", 60000)

print("Before Company Name Change")
employee1.show_Details()
employee2.show_Details()

# Changing the companyName
Employee.change_company("Green Tech PVT LTD")

print("After Company Name Change")
employee1.show_Details()
employee2.show_Details()

    
```
