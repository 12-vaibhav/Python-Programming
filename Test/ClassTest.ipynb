<h2>Answer: 2</h2>

```python
class Employee:
    # Class Variable
    CompanyName = "ABC PVT LTD"

    # Instance Variables
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary
    
    # Using Instance method to display employee data
    def show_Details(self):
        print(f"Name: {self.name} \n Salary: {self.salary} \n Company: {self.CompanyName}")

    @classmethod
    def change_company(cls, new_name):
        cls.CompanyName = new_name

# Creating two objects
employee1 = Employee("Alice", 50000)
employee2 = Employee("Bob", 60000)

print("Before Company Name Change")
employee1.show_Details()
employee2.show_Details()

# Changing the companyName
Employee.change_company("Green Tech PVT LTD")

print("After Company Name Change")
employee1.show_Details()
employee2.show_Details()

    
```

```python
class BankAccount:
    def __init__(self, name, initial_balance):
        # Public Variable
        self.name = name
        # Private Variable
        self.__balance__ = initial_balance

    def deposit(self, amount):
        if amount > 0:
            self.__balance__ += amount
            print(f"Deposited Amount: {amount}, New Balance: {self.__balance__}")
        else:
            print("Deposited amount must be positive.")

    def withdraw(self, amount):
        if amount > self.__balance__:
            print("Insufficient Balance")
        elif amount <= 0:
            print("Withdrawl amount must be positive")
        else:
            self.__balance__ -= amount
            print(f"Withdraw: {amount}, Remaining balance: {self.__balance__}")

    def showBalance(self):
        print(f"Account Holder: {self.name}")
        print(f"Current Balance: {self.__balance__}")

# Testing the Class

# Creating an object
account = BankAccount("John Doe", 1000)

# Testing showBalance
account.showBalance()

# Testing deposit
account.deposit(5000)

# Testing Withdrawl
account.withdraw(500)

# Testing Withdrawl insufficient balance
account.withdraw(15000)

```
