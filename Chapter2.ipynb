<h1>Strings</h1>
<h2>Initialisation</h2>

<ul>
    <li>Strings are immutable suquences of characters.</li>
</ul>

<dl>
    <dt>Created using: </dt>
    <dd>- Single Quotes: 'hello'</dd>
    <dd>- Double Quotes: "hello"</dd>

```python
s1 = 'Python'
s2 = "Data Science"
s3 = """Line1
Line2"""

print(s1)
print(s2)
print(s3)
```

<h2>Operators on strings</h2>

<ul>
    <li>+ concatenation, * repetition</li>
    <li>in membership</li>
    <li>comparison uses lexicographic order</li>
</ul>

```python
print("Py" + "thon")
print("Veer" * 4)
print("e" in "Veer")
print("Veer" > "Vaibhav")
```

```python
Name = ["Vaibhav", "Kumar"]
"Vaibhav" in Name
```

<h2>Indexing</h2>

```python
Name = "Vaibhav"
print(Name[0])
print(Name[-0])

print(Name[len(Name)-1])
```

<h2>Slicing</h2>
<p>Form: s[start : stop : step] (stop excluded)</p>

```python
Name = "Vaibhav"
print(Name[1:3])
print(Name[:3])
```

<h2>Split()</h2>
<p>Splits into list by delimiter (default = whitespace)</p>

```python
print("one two three".split())
print("12-07-2007".split("-"))
```
