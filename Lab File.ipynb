<h2 style="color:lightblue;">Built in Functions vs Methods (Internal Working)</h2>
<h2>Built in Functions (Global / Independent)</h2>
<p>Example: </p>

```python
len("Vaibhav")
```

<h2>Internal meaning: </h2>
<p>Python searches the len inside the built-in namespace.</p>

<p>Builts-ins live in: </p>
<br>

<ol>
    <li>__builtins__</li>
    <li>builtins module</li>
</ol>

<p>So, this is basically: <br><br> builtins.len("Vaibhav") <br> Built-in functions are not tied to one object.</p>

<br>
<hr>

<h2>Method (belongs to an object's class)</h2>
<p>Example: </p>

```python
"vaibhav".upper()
```

<h2>Internal meaning: </h2>
<p>Python does: </p>

<ol>
    <li>Look at the object type: string</li>
    <li>Find upper inside string</li>
    <li>Bind it with the object "vaibhav"</li>
    <li>then call it</li>
</ol>

<br>

<p>Equivalent internal call: <br> str.upper("vaibhav") <br> <br> So, method = function that becomes bound to an object.</p> <br>
